# Correlation Plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, date
```

#### Now we are going to import our data and convert it required form

```python
#data Import
location = "kolkata"

filename_t2 = "WD_{0}_{1}.csv"
filed = filename_t2.format(location, "D")

def apply_dt(string):
    return datetime.strptime(string, '%Y-%m-%d').date()

df = pd.read_csv(filed)

df['date'] = df['date'].apply(apply_dt)

del(df['Unnamed: 0'])
```

## Plots

```python
plt.scatter(df['date'], df['maxtempC'])
plt.show()

plt.scatter(df['date'], df['mintempC'])
plt.show()
```
